```python
import pandas as pd
```

```python
# Read the dataset
data = pd.read_csv('/Users/priyakundu/Documents/NYU Capstone WaterVue Files/Data Processing/Updated_Dataframe_WaterQual.csv')
data
```

```python
# Convert "Sample Date" to datetime if it's not already
data['Sample Date'] = pd.to_datetime(data['Sample Date'])

# Group data by Location
grouped = data.groupby('Location')

# Initialize a dictionary to store the results
results = {'Location': [], 'Mean Dissolved Oxygen': [], 'Dissolved Oxygen Category': []}

# Define functions to categorize dissolved oxygen levels
def categorize_oxygen_level(oxygen_level):
    if oxygen_level >= 5.0:
        return "Optimal"
    elif 2.1 <= oxygen_level <= 5.0:
        return "Biological Stress"
    elif 0.6 <= oxygen_level <= 2.0:
        return "Hypoxic"
    elif 0.0 <= oxygen_level <= 0.5:
        return "Anoxic"
    else:
        return "Unknown"

# Iterate through each group
for name, group in grouped:
    mean_oxygen = group['Dissolved Oxygen'].mean()
    category = categorize_oxygen_level(mean_oxygen)
    results['Location'].append(name)
    results['Mean Dissolved Oxygen'].append(mean_oxygen)
    results['Dissolved Oxygen Category'].append(category)

# Create a DataFrame from the results dictionary
result_df = pd.DataFrame(results)

# Sort the DataFrame by Dissolved Oxygen Category
result_df = result_df.sort_values(by='Dissolved Oxygen Category')
```
