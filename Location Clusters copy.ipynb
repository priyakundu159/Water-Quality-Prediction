```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Read the dataset
data = pd.read_csv('/Users/priyakundu/Documents/NYU Capstone WaterVue Files/Data Processing/Updated_Dataframe_WaterQual.csv')
data
```

```python
# Select the numerical columns for clustering
X = data.iloc[:, 2:].values

# Normalize the data
scaler = StandardScaler()
data_normalized = scaler.fit_transform(X)

# Define the number of clusters
num_clusters = 2

# Perform KMeans clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(data_normalized)

# Assign cluster labels to locations
location_clusters = {}
for location, cluster in zip(data["Location"].unique(), clusters):  # Assuming you have a list of locations
    if cluster not in location_clusters:
        location_clusters[cluster] = [location]
    else:
        location_clusters[cluster].append(location)

# Print out the locations in each cluster
for cluster, locations in location_clusters.items():
    print(f"Cluster {cluster + 1}:")
    print(locations)

# Analyze the cluster centroids and parameter ranges
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
for i, center in enumerate(cluster_centers):
    print(f"\nCluster {i + 1} Centroid (parameter means):")
    print(center)

    # Find minimum and maximum values for each parameter in the cluster
    cluster_data = data[clusters == i]  # Get data points belonging to this cluster
    parameter_mins = np.min(cluster_data, axis=0)
    parameter_maxs = np.max(cluster_data, axis=0)

    print(f"Min values for parameters:")
    print(parameter_mins)
    print(f"Max values for parameters:")
    print(parameter_maxs)
```
